```python
import beauty
import csv
import pandas
import matplotlib
import matplotlib.pyplot as plt
#matplotlib.style.use('ggplot')
%matplotlib inline
beauty.set_tex(False)
pandas.set_option('display.max_colwidth', -1)
```

```python
df = pandas.DataFrame.from_csv('DATA/coded_ALL.csv')
```

```python
df.columns
```

```python
df[(df.trust == -1)&(df.gender == 'male')].candidate.value_counts().plot(kind='bar')
plt.show()
```

```python
df[(df.trust == -1)&(df.gender == 'female')].candidate.value_counts().plot(kind='bar')
plt.show()
```

```python
df[(df.trust == -1)&(df.group == 'A')].org.value_counts().plot(kind='bar')
plt.title('Show Source Distrust')
plt.show()
df[(df.trust == -1)&(df.group == 'B')].org.value_counts().plot(kind='bar')
plt.title('Don\'t Show Source Distrust')
plt.show()
```

```python
df[(df.trust == -1)&(df.group == 'A')].candidate.value_counts().plot(kind='bar')
plt.title('Show Source Distrust')
plt.show()
df[(df.trust == -1)&(df.group == 'B')].candidate.value_counts().plot(kind='bar')
plt.title('Don\'t Show Source Distrust')
plt.show()
```

```python
CANDIDATES = set(df.candidate)
for c in CANDIDATES:
    df[(df.trust == -1)&(df.group == 'A')&(df.candidate==c)].org.value_counts().plot(kind='bar')
    plt.title("Untrusted Upon Show For " + c)
    plt.show()
```

```python
df[(df.group == 'A')&(df.candidate=="Hillary Clinton")].org.value_counts().plot(kind='bar')
plt.show()
```

## any effect on polar ends?

```python
print df.flesch[df.flesch <6].count()
print df.flesch[df.flesch > 12].count()
```

```python
df[df.flesch <6][:100].trust.value_counts().sort_index().plot(kind='bar')
plt.ylim([0,50])
plt.title("Low RL")
```

```python
df[df.flesch > 12][:100].trust.value_counts().sort_index().plot(kind='bar')
plt.ylim([0,50])
plt.title("Hi RL")
```

```python
from scipy.stats import ttest_ind
from scipy.stats import wilcoxon
from scipy.stats import chisquare
```

```python
ttest_ind(df[df.flesch < 6].trust,df[df.flesch > 12].trust)
```

```python
# we don't really have paired samples though
wilcoxon(df[df.flesch < 6].trust[:100], df[df.flesch > 12][:100].trust)
```

```python
wilcoxon(df[df.flesch < 6].trust[:100], df[df.flesch > 12][:100].trust)
```
