```python
import beauty
from textstat.textstat import textstat
import csv
import pandas
import matplotlib
import matplotlib.pyplot as plt
#matplotlib.style.use('ggplot')
%matplotlib inline
beauty.set_tex(False)
```

```python
group_A = pandas.DataFrame.from_csv('DATA/demographics_A.csv')
group_B = pandas.DataFrame.from_csv('DATA/demographics_B.csv')
group_AB = pandas.concat([group_A, group_B])
```

```python
group_A.columns
group_B.columns
```

## Demographics (Gender, Age, Location)

```python
group_A['what_is_your_gender'].value_counts().plot.pie(label="", autopct='%1.1f%%', fontsize=20, figsize=(6, 6))
plt.ylabel('Gender Breakdown for Group A', fontsize=20)
plt.savefig('FIGS/gender_A.png')
plt.show()

group_B['what_is_your_gender'].value_counts().plot.pie(label="", autopct='%1.1f%%', fontsize=20, figsize=(6, 6))
plt.ylabel('Gender Breakdown for Group B', fontsize=20)
plt.savefig('FIGS/gender_B.png')
plt.show()
```

```python
group_A[['_city','_region']].to_csv('DATA/locations_A.csv')
group_B[['_city','_region']].to_csv('DATA/locations_B.csv') 
```

```python
group_A['_region'].value_counts().plot(kind='bar',fontsize=3)
plt.title('Participants by State Group A',fontsize=5)
plt.savefig('FIGS/state_A.png')
plt.show()

group_B['_region'].value_counts().plot(kind='bar',fontsize=3)
plt.title('Participants by State Group B',fontsize=5)
plt.savefig('FIGS/state_B.png')
plt.show()
```

```python
#AGE GROUP
age_to_pretty = {1829:'18-29', 3049:'30-49', 5064:'50-64',64:'64+'}
group_A.age_group = group_A.what_is_your_age.apply(lambda row: age_to_pretty[row])
group_B.age_group = group_B.what_is_your_age.apply(lambda row: age_to_pretty[row])
group_AB.age_group = group_AB.what_is_your_age.apply(lambda row: age_to_pretty[row])

group_A.age_group.value_counts().plot(kind='bar',fontsize=7)
plt.title('Age Groups, Group A', fontsize=7)
plt.tight_layout()
plt.savefig('FIGS/ages_A.png')
plt.show()

group_B.age_group.value_counts().plot(kind='bar',fontsize=7)
plt.title('Age Groups, Group B', fontsize=7)
plt.tight_layout()
plt.savefig('FIGS/ages_B.png')
plt.show()
```

```python
beauty.set_tex=False
group_A.what_is_your_political_party_affiliation.value_counts().plot(kind='bar', fontsize=6)
plt.tight_layout()
plt.savefig('FIGS/party_A.png')
plt.show()

group_B.what_is_your_political_party_affiliation.value_counts().plot(kind='bar', fontsize=6)
plt.tight_layout()
plt.savefig('FIGS/party_B.png')
plt.show()
```

## Voting preference by region

```python
states = group_AB._region.value_counts()[group_AB._region.value_counts() > 5].index
for state in states:
    group_AB[group_AB._region==state].if_the_election_were_tomorrow_i_would_vote_for.value_counts().plot(kind='bar', fontsize=5)
    plt.title('Preferred Candidate, ' + state + ', All', fontsize=7)
    plt.tight_layout()
    plt.savefig('FIGS/voting_for_AB_' + state + '.png')
    plt.show()
```
