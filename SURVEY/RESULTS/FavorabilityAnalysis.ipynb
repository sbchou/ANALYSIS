```python
import beauty
import csv
import pandas
import matplotlib
import matplotlib.pyplot as plt
#matplotlib.style.use('ggplot')
%matplotlib inline
beauty.set_tex(False)
```

```python
group_A = pandas.DataFrame.from_csv('DATA/flattened_A.csv')
group_B = pandas.DataFrame.from_csv('DATA/flattened_B.csv')
group_A['group'] = 'A'
group_B['group'] = 'B'
group_AB = pandas.concat([group_A, group_B])
group_AB.to_csv('DATA/flattened_ALL.csv')
```

```python
#len(group_A[(group_A.candidate=='Hillary Clinton') & (group_A.favor > 0)])
len(group_AB)
```

## Overall favorability by candidate

```python
score_to_value = {-1:'Unfavorable', 0:'Neutral', 1:'Positive'}
group_AB.favor_score = group_AB.favor.copy()
group_AB.favor = group_AB.favor_score.apply(lambda row:score_to_value[row])
```

```python
CANDIDATES = set(group_AB.candidate)
 
for c in CANDIDATES:
    group_AB[group_AB.candidate == c]['favor'].value_counts().sort_index().plot(kind='bar')
    plt.tick_params(axis='both', which='major', labelsize=5)
    plt.title(c.split()[-1] + " Media Favorability Score", fontsize=7)
    #plt.xticks([1,2,3],[1,2,3])
    plt.ylim([0,100])
    plt.tight_layout()
    plt.savefig('FIGS/favorability/' + c.split()[-1] + '_favorability.png')
    plt.show()
 
```

```python
# Get favorability percentage
for c in CANDIDATES:
    print c
    print group_AB[group_AB.candidate==c].favor.value_counts(normalize=True).sort_values() * 100
    print
```

## What is the same breakdown for trustworthy stories?

```python
for c in CANDIDATES:
    group_AB[(group_AB.candidate == c) & (group_AB.trust > 0)]['favor'].value_counts().sort_index().plot(kind='bar')
    plt.tick_params(axis='both', which='major', labelsize=5)
    plt.title(c.split()[-1] + " Media Favorability Score", fontsize=7)
    #plt.xticks([1,2,3],[1,2,3])
    plt.ylim([0,100])
    plt.tight_layout()
    plt.savefig('FIGS/favorability/' + c.split()[-1] + '_favorability_trust_gt_0.png')
    plt.show()
 
```

## Overall trustworthiness by candidate

```python
trust_to_value = {2:'Strongly Agree', 1:'Agree', 0:'Neutral', -1:'Disagree', -2:'Strongly Disagree'}
group_AB.trust_score = group_AB.trust.copy()
group_AB.trust = group_AB.trust_score.apply(lambda row:trust_to_value[row])
```

```python
CANDIDATES = set(group_AB.candidate)
 
for c in CANDIDATES:
    group_AB[group_AB.candidate == c]['trust'].value_counts().plot(kind='bar')
    plt.tick_params(axis='both', which='major', labelsize=5)
    plt.title("Articles About " + c.split()[-1] + " Trustworthy", fontsize=7)
    #plt.xticks([1,2,3],[1,2,3])
    plt.ylim([0,100])
    plt.tight_layout()
    plt.savefig('FIGS/trustworthiness/' + c.split()[-1] + '_trustworthiness.png')
    plt.show()
 
```

```python
# Get favorability percentage
for c in CANDIDATES:
    print c
    print group_AB[group_AB.candidate==c].trust.value_counts(normalize=True).sort_index() * 100
    print
```

## Trust to binary

```python
trust_to_binary_value = {2:'Agree', 1:'Agree', 0:'Neutral', -1:'Disagree', -2:'Disagree'}
group_AB['trust_binary'] = group_AB.trust_score.apply(lambda row:trust_to_binary_value[row])
```

```python
CANDIDATES = set(group_AB.candidate)
 
for c in CANDIDATES:
    group_AB[group_AB.candidate == c]['trust_binary'].value_counts().plot(kind='bar')
    plt.tick_params(axis='both', which='major', labelsize=5)
    plt.title("Articles About " + c.split()[-1] + " Trustworthy", fontsize=7)
    #plt.xticks([1,2,3],[1,2,3])
    plt.ylim([0,100])
    plt.tight_layout()
    plt.savefig('FIGS/trustworthiness/' + c.split()[-1] + '_trustworthiness_binary.png')
    plt.show()
 
```

```python
# Get trust percentage
for c in CANDIDATES:
    print c
    print group_AB[group_AB.candidate==c].trust_binary.value_counts(normalize=True).sort_index() * 100
    print
```
